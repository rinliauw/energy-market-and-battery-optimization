## 1. solar exposure data
http://www.bom.gov.au/jsp/ncc/cdio/weatherData/av?p_nccObsCode=193&p_display_type=dailyDataFile&p_startYear=&p_c=&p_stn_num=086095

```python
# join solar external data to market data
import pandas as pd

solar_data = pd.read_csv('../../raw_data/solar_exposure_melbourne.csv')
market_data = pd.read_excel('../../raw_data/market_data.xlsx')
```

```python
date = pd.to_datetime(solar_data[['Year', 'Month', 'Day']])
```

```python
solar_data['Date'] = date
```

```python
solar_data = solar_data.drop(columns=['Year', 'Month', 'Day'])
```

```python
solar_data = solar_data[['Date', 'Daily global solar exposure (MJ/m*m)']]
```

```python
solar_data['Date'] = solar_data['Date'].astype('datetime64[ns]')
```

```python
market_data['Time'] = market_data['Time (UTC+10)'].dt.date
market_data['Time'] = market_data['Time'].astype('datetime64[ns]')
```

```python
solar_merge = pd.merge(market_data, solar_data, left_on = 'Time', right_on = 'Date', how = 'left')
```

```python
solar_merge = solar_merge.drop(columns='Date')
```

```python
solar_merge
```

## 2. covid cases data

```python
# join covid cases data to market data
covid_cases = pd.read_excel('../../raw_data/covid_cases_vic.xlsx')
```

## 3. extreme weather
The AER’s State of the Energy Market 2020 report notes that extreme weather caused price spikes in early 2020, when bushfires caused transmission lines to trip and limited generation. These transmission interruptions led to market separation between the regions of the NEM, with spot prices hitting the then-cap of $14,700 per MWh on multiple days while the bushfires raged.

In 2019, volatility occurred in Victoria, South Australia and Tasmania, linked to extreme weather and high system demand; an unplanned outage at Loy Yang A Power Station that ran for several months and removed 11 per cent of low cost generation from the region; and outages at the Yallourn and Mortlake power stations that compounded the situation.

In 2018, an oppressive heatwave coincided with a crash at Loy Yang B Power Station that took about 528 megawatts of energy out of Victoria’s state grid. On the day of the crash, wholesale prices in Victoria jumped from 86.72 at 3:30pm to 12,931, before sliding back down to $238.45 per MWh by 6:30pm.

On the other hand, market volatility can also be seen in an increasing occurrence of negative prices (at these times, generators pay to stay connected to the grid when demand is low, if it comes at a lower cost than shutting down and re-starting their plants). The market set a record number of negative prices in the second half of 2019, typically at times when weather conditions are optimal for solar and wind generation, leading to oversupply.

```python
temp_2018 = pd.read_csv("../../raw_data/melbourne_temp_2018.csv")
temp_2019 = pd.read_csv("../../raw_data/melbourne_temp_2019.csv")
temp_2020 = pd.read_csv("../../raw_data/melbourne_temp_2020.csv")
temp_2021 = pd.read_csv("../../raw_data/melbourne_temp_2021.csv")

# join all
temp_all_year = pd.concat([temp_2018, temp_2019, temp_2020, temp_2021])
temp_all_year['date'] = pd.to_datetime(temp_all_year[['Year','Month','Day']])
```

```python
temp_all_year = temp_all_year[['date', 'Maximum temperature (Degree C)']]
temp_merge = pd.merge(covid_merge, temp_all_year, left_on = "Date", right_on = "date")
temp_merge.drop(columns = ['date', 'Date'])
```

```python
# export to csv
temp_merge.to_csv('../../preprocessed_data/external_data_joined.csv')
```
